```python
# Combien de lignes sont exactement les mêmes ?
exact_duplicates = df[df.duplicated()]
print("Exact duplicates:", exact_duplicates.shape)

# Combien de lignes sont similaires sur certaines colonnes ?
partial_duplicates = df[df.duplicated(subset=["age", "sex", "cp", "chol", "thal"])]
print("Partial duplicates:", partial_duplicates.shape)
```

```python
# Afficher les 5 premiers vrais doublons détectés
exact_duplicates = df[df.duplicated()]
print("Nombre de doublons parfaits détectés :", exact_duplicates.shape[0])
exact_duplicates.head(5)
```

```python
# Trouver et afficher toutes les lignes qui sont des doublons parfaits (sauf la première occurrence)
duplicates = df[df.duplicated(keep=False)]  # keep=False pour afficher toutes les occurrences des doublons
print("Nombre total de doublons parfaits (incluant premières occurrences):", duplicates.shape[0])

# Affiche les 5 premiers doublons
print(duplicates.head(5))
```

```python
# Exemple : afficher les doublons pour une ligne spécifique (ici index 0)
print(df.iloc[0])
print(df[df.eq(df.iloc[0]).all(axis=1)])  # Toutes les lignes identiques à la ligne 0
```

```python
duplicates = df[df.duplicated(keep=False)]
print(duplicates.head(10))
```

```python
indices_a_tester = [0, 31, 43, 55, 61]  # Par exemple, tu peux changer cette liste

for idx in indices_a_tester:
    ligne_test = df.iloc[idx]
    doublons_precis = df[df.eq(ligne_test).all(axis=1)]
    print(f"\nLigne d'index {idx} a {doublons_precis.shape[0]} occurrence(s) identique(s) :")
    print(doublons_precis)
```
